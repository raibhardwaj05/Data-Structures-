## Patterns

```python
# time complexity = O(6 x 5) = O(N**2)
for i in range(6):
    for j in range(5):
        print("* ", end="")
    print() # for new line
```

```python
# time complexity = O(36 / 2) = O(N**2 / 2) = O(N**2)

for i in range(6):
    for j in range(i + 1):
        print("* ", end = "")
    print()
```

```python
# time complexity = O(N**2 / 2)

for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end= "")
    print()
```

```python
# time complexity = O(N**2 / 2)

for i in range(5):
    for j in range(5 - i):
        print("* ", end = "")
    print()
```

```python
# time complexity = O(N**2 / 2)

for i in range(5):
    for j in range(1, 6 - i):
        print(j, end = "")
    print()
```
