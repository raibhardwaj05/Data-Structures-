## Patterns

```python
# time complexity = O(6 x 5) = O(N**2)
for i in range(6):
    for j in range(5):
        print("* ", end="")
    print() # for new line
```

```python
# time complexity = O(36 / 2) = O(N**2 / 2) = O(N**2)

for i in range(6):
    for j in range(i + 1):
        print("* ", end = "")
    print()
```

```python
# time complexity = O(N**2 / 2)

for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end= "")
    print()
```

```python
# time complexity = O(N**2 / 2)

for i in range(5):
    for j in range(5 - i):
        print("* ", end = "")
    print()
```

```python
# time complexity = O(N**2 / 2)

for i in range(5):
    for j in range(1, 6 - i):
        print(j, end = "")
    print()
```

```python
# time complexity = O(N**2)

# Loop checks → O(N) 
# Printing → O(N²) ==>> 1+3+5+⋯+(2N−1)=N**2
# The dominant term is O(N²).

for i in range(5):
    print(" "*(4 - i) + "*"*(2*i + 1))
```

```python
# time complexity = O(N**2)

for i in range(5):
   print(" "*i + "*"*((2*5) -1 - (2*i))) 
```

```python
# time complexity = O(N**2) + O(N**2) = O(N**2)

for i in range(5):
    print(" "*(4 - i) + "*"*(2*i + 1))

for i in range(5):
   print(" "*i + "*"*((2*5) -1 - (2*i)))
```

```python
# time complexity = O(N**2)
# space complexity = O(1)

n = 5

for i in range(2*n):
    if i > n: 
        i = 9 - i
    for j in range(1, i + 2):
       print("*", end = "")

    print()
```

```python
# time compexity = O(N**2 / 2) ==> if - else doesn't contribute much in the time complexity so...

for i in range(5):
    for j in range(1, i + 2):
        if (i + j) % 2:
            print("1 ", end = "")
        else:
            print("0 ", end = "")

    print()
```

```python
# time complexity = O(N**2 / 2)+O(N**2)+O(N**2 / 2)

N = 5

space = 2*(N - 1)

for i in range(N):
    for j in range(1, i + 2):
        print(j, end="")

    for j in range(space):
        print(" ", end="")

    for j in range(i+1, 0, -1):
        print(j, end="")

    space -= 2
    print()
```
